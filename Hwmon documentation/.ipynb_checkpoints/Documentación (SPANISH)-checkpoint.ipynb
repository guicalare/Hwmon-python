## Motivaciones para crear Hwmon

Hwmon se ha creado con la intencin de reemplazar libreras y APIs de Linux con las cuales obtener informacin del sistema sin necesidad de depender de dependencias ajenas a un sistema Linux estndar, es decir, que no se necesite instalar nada.

Los nicos requisitos para ejecutar Hwmon en un sistema son:

- El SO sea Linux
- Disponer de python 3

Hwmon es adems un librera desarrollada expresamente por y para Python 3 con funciones sencillas de entender y de funcionar, las cuales leen y sintetizan en una misma librera toda la informacin til para monitorizar un sistema Linux que se encuentran en las carpetas /sys, /proc y /dev. Por lo que solo se obtendr la informacin que el propio sistema tenga registrada.

## Por que usar Hwmon en vez de otras libreras?

Como ya se ha mencionado, Hwmon no requiere de dependencias ni programas y es una librera creada con las libreras estndares de Python 3. Cosa que no ocurre con otras libreras como son:

- Pysensors
- lm-sensors
- psutil

Donde se necesitan de archivos y programas para poder funcionar, de forma que si te falta algo de eso, simplemente no funciona. Adems de son soluciones mas pesadas en cuanto a tamaos de archivos que Hwmon.

## Es realmente una alternativa viable a lm-sensors?

S. Hwmon es capaz de retornar e imprimir la misma informacin que se obtiene al instalar y ejecutar sensors. 



E incluso imprime por pantalla la informacin de una forma bonita y amigable de forma que sea mas til y fcil de entender.

## Que informacin puedo obtener con Hwmon y de donde la obtiene?

Hwmon es capaz de extraer:

- Informacin de los sensores disponibles en el ordenador
- Informacin del procesador
- Informacin de la memoria del sistema
- Informacin de los paquetes enviados y recibidos
- Dispositivos USB conectados al ordenador
- Discos conectados al ordenador

Toda esa informacin se extrae respectivamente de los siguientes sitios:

- /sys/class/hwmon
- /proc/cpuinfo y /proc/stat
- /proc/meminfo
- /proc/net/dev
- /dev/input/by-id
- /dev/disk/by-id

## Como usar y llamar a la librera?

```python
from hwmon import Hwmon
```

# Informacin de los sensores

Para acceder a la subclase de los sensores, se ejecutara la siguiente sentencia:


```python
sensors = Hwmon.HW()
```

Siendo las funciones que nos interesan en este caso: data() y print_data() . Veamos un ejemplo de cada funcin:

### Imprimir sensores

Esta funcin imprimir de forma tabulada y bonita toda la informacin de los sensores del sistema.

```python
sensors.print_data()
```

### Obtener datos

En el caso de que nos interese obtener la informacin de los sensores en vez de imprimirlos por pantalla, ejecutaremos la siguiente sentencia:

```python
sensors.data()
```
